```python
# Importando biblioteca
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pdfplumber
import nltk
```

```python
# Carregando o arquivo
pdfMetier = pdfplumber.open('/content/RespostasMetier.pdf')
```

```python
# Extraindo as informações dos textos
paginas = pdfMetier.pages[0]
textoCRU = paginas.extract_text()

# Mostrando resuldato
textoCRU
```

```python
# Quebrando o texto em palavras individuais
nltk.download('punkt')
lista_de_palavras = nltk.tokenize.word_tokenize(textoCRU)
lista_de_palavras
```

```python
# Transformando todas as palavra em minúsculas
lista_de_palavras = [palavra.lower() for palavra in lista_de_palavras]
lista_de_palavras
```

```python
# Criando uma lista que contêm pontuação que desejamos remover
pontuacao = ['(', ')', ';', ':', '[', ']', ',', '.', '?', '/', 'x', "''", "``"]
```

```python
# Criando uma lista de stop word
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('portuguese')
stop_words
```

```python
# Criando uma lista de palavras sem stopword e pontuações
keywords = [palavra 
            for palavra in lista_de_palavras 
            if not palavra in stop_words and not palavra in pontuacao]
keywords
```

```python
# Contando as palavras
len(keywords)
```

```python
# Concatenando as palavras
textoMetier = ' '.join(s for s in keywords)
textoMetier
```

# Gerando Núvem de Palavras

```python
# Núvem de Palavras de todas as perguntas
wordcloud = WordCloud(max_font_size = 150,
                      width = 1280,
                      height = 720).generate(textoMetier)

fig, ax = plt.subplots(figsize = (16, 9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)
plt.show()
```
